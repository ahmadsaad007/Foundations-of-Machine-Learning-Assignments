```python

def predict(dict1, labels):
    wrongpredict = 0
    for i in dict1.keys():
        dictValues={}
        dictValues = dict1.get(i)
        maxValue = max(dictValues.values())
        sumValues = sum(dictValues.values())
        wrongpredict += sumValues-maxValue
    print("Training Error:", (wrongpredict/len(labels))*100.0,"%")

    return
```

```python

def main():
    
    '''
    Get the first command line argument of the program.
    For example, sys.argv[1] could be a string such as 'breast_cancer.csv' or 'titanic_train.csv'
    '''
    #szDatasetPath = sys.argv[1]
	# Comment out the following line and uncomment the above line in your final submission
    #szDatasetPath = 'breast_cancer.csv'
    szDatasetPath = 'titanic_train.csv'

    '''
    Read the data from the csv file
    listColNames[j] stores the jth column name
    listData[i][:-1] are the features of the ith example
    listData[i][-1] is the target value of the ith example
    '''
    listColNames = [] # The list of column names
    listData = [] # The list of feature vectors of all the examples
    nRow = 0
    with open(szDatasetPath) as csvFile:
        csvReader = csv.reader(csvFile, delimiter=',')
        for row in csvReader:
            if 0 == nRow:
                listColNames = row
            else:
                listData.append(row)
            nRow += 1

    '''
    Scan the data and store the unique values of each column.
    listColUniqueVals[j] stores a list of unique values of the jth column
    '''
    
    Y = [] #Stores the class labels for every example
    listDataEncoded = [] #The list of feature vectors except the class labesls of all the examples 
    
    listColUniqueVals = [[] for i in range(len(listColNames))]
    for example in listData:
        Y.append(example[-1])
        tempList = []
        for i in range(len(example)):
            if i < (len(example)-1):
                tempList.append(example[i])
            if example[i] not in listColUniqueVals[i]:
                listColUniqueVals[i].append(example[i])
        listDataEncoded.append(tempList)
        
    # For each feature, compute the training error of a one-level decision tree
    for i in range(len(listColNames)-1):
        dictionary= {}
        for example in listData:
            featureValue = example[i]
            if(featureValue in dictionary):
                dictValues={}
                dictValues = dictionary.get(featureValue)
                #get the inner dictionary
                #Now try to find the corresponding label
                if(example[-1] in dictValues.keys()):
                    dictValues[example[-1]] += 1
                else:
                    dictValues[example[-1]] = 1
            else:
                dictValues={}
                dictValues[example[-1]] = 1
                dictionary[featureValue] = dictValues
   
        #make a prediction by calling predict method. Method gets arguments dict and label
        #and then calls the training error
        print("Feature: ", listColNames[i])
        predict(dictionary, Y)
    
    for i in range(len(listColNames)-1):
        le = LabelEncoder()
        le.fit(listColUniqueVals[i])
        featureValue =[]
        for j,example in enumerate(listData,0):
            featureValue.append(example[i])
        featureValue = le.transform(featureValue)
        for j in range(len(listData)):
            listDataEncoded[j][i] = featureValue[j]
            
    X = listDataEncoded
    lle = LabelEncoder()
    lle.fit(listColUniqueVals[-1])
    y = lle.transform(Y)
    
    clt = DecisionTreeClassifier()
    clt.fit(X,y)
    print("Decision Tree Training Error:", (1 - clt.score(X, y))*100.0, "%")
    return None
          
          
if __name__ == '__main__':

    main()


```
